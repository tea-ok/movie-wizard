## Inspecting the data

```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
df = pd.read_csv('data.csv', na_values='\\N')
df = df.dropna(subset=['genres'])
df.head()
```

```python
df['genres'] = df['genres'].str.split(',')

# Use explode to transform each element of a list-like to a row, and find unique genres
df_genres = df.explode('genres')
unique_genres = df_genres['genres'].unique()
unique_genres
```
